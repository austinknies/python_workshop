```python
def get_fv(row): # Fundamental value
    ev = 24
    remaining_pmts = 10-row['period']+1 # 10 total payments, pay out at end of period
    result = remaining_pmts * ev
    return result
```

```python
df['fv'] = df.apply(get_fv, axis=1) # runs function 200 times with axis = 1 (each row)
df
```
